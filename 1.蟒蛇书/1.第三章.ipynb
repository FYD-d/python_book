```python
# 8
# (1)
place=['China','America','England','France','Russia']

print(f'I want to go to {place}')

place_sorted=sorted(place)
print(place)
print(place_sorted)

place_sorted2=sorted(place,reverse=True)
print(place)
print(place_sorted2)
```

```python
# (2)
place=['China','America','England','France','Russia']
place.reverse()
print(place)

place.reverse()
print(place)
```

```python
# (3)
place=['China','America','England','France','Russia']
place.sort()
print(place)

place.sort(reverse=True)
print(place)
```
