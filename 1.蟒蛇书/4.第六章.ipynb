```python
# 5
rivers={'nile':'egypt','long_river':'china','ocean':'england'}
for river,country in rivers.items():
    print(f'\n{river} runs through {country}')

for river in rivers.keys():
    print(river)

for country in rivers.values():
    print(country)
```

```python
# 6
favourite_languages={
    'jen':'python',
    'sarah':'c',
    'edward':'ruby',
    'phil':'python'
}

check_people=['jen','sarah','peter']

for person in check_people:
    if person not in favourite_languages.keys():
        print(f'Invite you {person.title()}' )
    else:
        print(f'Thank you {person.title()}')


for person in set(favourite_languages.values()):
    print(person)
```

```python
# 11
cities={
    '上海':{
        'country':'中国',
        'population':'1千万'
    },
    '北京':{
        'country':'中国',
        'population':'2千万'
    },
    '莫斯科':{
        'country':'俄罗斯',
        'population':'3亿'
    }
}

for city,factor in cities.items():
    print(f'\ncity_name:{city}')
    country=factor['country']
    population=factor['population']
    print(f'\tcountry:{country}')
    print(f'\tpopulation:{population}')
```

```python
# 12
favourite_languages={
    'jen':['python'],
    'sarah':['C'],
    'edward':['ruby','python'],
    'phil':['python','C++','Mysql']
}

for name,like_language in favourite_languages.items():
        if len(like_language)>1:
            print(f'\n{name.title()} likes many languages:')
            for language in like_language:
               print(f'\t{language}')
        else:
            language=like_language[0]
            print(f'\n{name.title()} only likes {language}')
```
