```python
# 5
while True:
    age=int(input('输入年龄：'))
    if age<3:
        print('免费')
        break
    elif age>=3 and age<12:
        print('10')
        break
    else:
        print('15')
        break
```

```python
# 8.
sandwich_orders=['水果','火腿','培根']
finished_orders=[]
while sandwich_orders:
    sandwich_order=sandwich_orders.pop()
    print(f'I made your {sandwich_order} sandwich')
    finished_orders.append(sandwich_order)
for finish in finished_orders:
    print(finish)
```

```python
# 9.
sandwich_orders=['水果','火腿','培根','pastrami','pastrami']
while 'pastrami' in sandwich_orders:
    sandwich_orders.remove('pastrami')
print(sandwich_orders)
```
