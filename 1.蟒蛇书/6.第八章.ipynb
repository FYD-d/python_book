```python
# 8
def make_album(name,album):
    sentence=f'{name} writes {album}'
    return sentence

while True:
    print('enter q at any time to quit')
    name=input('tell me your name:')
    if name=='q':
        break
    album=input('tell album:')
    if album=='q':
        break
    sentence=make_album(name,album)
    print(sentence)
```

```python
# 10
def send_message(unsend_lists,send_lists):
    while unsend_lists:
        name=unsend_lists.pop()
        print(f'hello {name.title()}')
        send_lists.append(name)
def sent_message(send_lists):
    for send_list in send_lists:
        print(f'{send_list.title()} 已发送')

unsend_lists=['jack','harry','peter']
send_lists=[]
send_message(unsend_lists,send_lists)
sent_message(send_lists)
```

```python
# 11
def send_message(unsend_lists,send_lists):
    while unsend_lists:
        name=unsend_lists.pop()
        print(f'hello {name.title()}')
        send_lists.append(name)
def sent_message(send_lists):
    for send_list in send_lists:
        print(f'{send_list.title()} 已发送')
    for unsend_list in unsend_lists:
        print(f'{unsend_list.title()} ')
unsend_lists=['jack','harry','peter']
send_lists=[]
send_message(unsend_lists[:],send_lists)
sent_message(send_lists)
```

```python
# 12
def sandwich(size,*materials):
    print(size)
    for material in materials:
        print(f'-{material}')
sandwich(10,'牛肉','cabbage')
sandwich(23,'羊肉','水果','哈密瓜')
```

```python
# 13
def build_profile(first,last,**profile):
    profile['firstname'] = first
    profile['lastname'] = last
    return profile
profile=build_profile(first='fu',last='yd',location='Shanghai',gender='male')
print(profile)
```

```python
# 14
def car_info(creator,size,**profiles):
    profiles['creator']=creator
    profiles['size']=size
    return profiles
profiles=car_info('subaru','20',color='blue',tow_package=True)
print(profiles)
```
