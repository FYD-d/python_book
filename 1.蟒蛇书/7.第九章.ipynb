```python
# 6
class Restaurant:
    def __init__(self,restaurant_name,age):
        self.restaurant_name=restaurant_name
        self.age=age

    def describe_restaurant(self):
        print(f'{self.restaurant_name} 经营 {self.age} 年')
    def open_restaurant(self):
        print(f'{self.restaurant_name} is open!')

class IceCreamStand(Restaurant):
    def __init__(self,restaurant_name,age,flavours):
        super().__init__(restaurant_name,age)
        self.flavours=flavours
    def describe_flavours(self):
        print(f'{self.restaurant_name}有')
        for flavour in self.flavours:
            print(f'-{flavour}口味')
type=['草莓','西瓜']
IceCreamStand=IceCreamStand('哇哈哈','11',type)
IceCreamStand.describe_flavours()
```

```python
# 8
class User:
    def __init__(self,name,id,email):
        self.name = name
        self.id=id
        self.email=email
    def describe_user(self):
        print(f'名字：{self.name} '
              f'账户{self.id} '
              f'邮箱：{self.email}')
    def greet_user(self):
        print(f'hello {self.name}')

privileges_list=['add post', 'delete post']
class priviligees:
    def __init__(self,name,privileges_list=privileges_list):
        self.privileges_list=privileges_list
        self.name=name

    def show_priviligees(self):
        print(f'{self.name} 有这些权限：')
        for priviligee in self.privileges_list:
            print(f'- can {priviligee}')

class Admin(User):
    def __init__(self,name,id,email):
        super().__init__(name,id,email)
        self.priviligees=priviligees(name)

admin=Admin('Peter','123','fu@qq.com')
admin.priviligees.show_priviligees()
```

```python
# 9.
class Car:
    def __init__(self,brand,model):
        self.brand=brand
        self.model=model

class Battery:
    def __init__(self,battery=40):
        self.battery=battery
    def upgrade_battery(self):
        if self.battery!=65:
            self.battery=65
            print(f"电池已升级！当前容量：{self.battery}")
        else:
            print(f"电池已是最大容量：{self.battery}")

    def get_range(self):
        if self.battery==40:
            range_km=150
        elif self.battery==65:
            range_km=225
        print(f'能开{range_km}km')

class ElectricCar(Car):
    def __init__(self,brand,model,battery):
        super().__init__(brand,model)
        self.battery=Battery(battery)

car1=ElectricCar('Ford','Mustang',65)
car1.battery.upgrade_battery()
```
